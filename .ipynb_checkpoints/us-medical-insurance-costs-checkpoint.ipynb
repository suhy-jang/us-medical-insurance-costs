# U.S. Medical Insurance Costs

```python
import csv

class Patient:    
    def __init__(self, info):
        self.age = info['age']
        self.sex = info['sex']
        self.bmi = info['bmi']
        self.children = info['children']
        self.smoker = info['smoker']
        self.region = info['region']
        self.charges = info['charges']
        
def analysis(dataset):
    age_sum = 0
    regions = []
    cost_smoker = 0
    cost_non_smoker = 0
    age_sum_with_children = 0
    line = 0
    for row in dataset:
        if line == 0:
            line += 1
            continue
        patient = Patient(row)
        age_sum += int(patient.age)
        regions.append(patient.region)
        if patient.smoker == 'yes':
            cost_smoker += float(patient.charges)
        else:
            cost_non_smoker += float(patient.charges)
        if int(patient.children) > 0:
            age_sum_with_children += int(patient.age)
        line += 1
    avg_age = round(age_sum/line, 1)
    regions.sort(reverse=True)
    major_region = regions[0]
    cost_smoker = round(cost_smoker/line, 1)
    cost_non_smoker = round(cost_non_smoker/line, 1)
    avg_age_with_children = round(age_sum_with_children/line, 1)
    print ('average age: ', avg_age)
    print ('major region: ', major_region)
    print (f"cost of smokers: {cost_smoker}, cost of non smokers: {cost_non_smoker}")
    print ('average age who has children: ', avg_age_with_children)

    
with open('insurance.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    analysis(reader)

```
