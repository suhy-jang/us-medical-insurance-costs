# U.S. Medical Insurance Costs

```python
import csv

age_sum = 0
line = 0
regions = []
cost_smoker = 0
cost_non_smoker = 0
age_sum_with_children = 0
with open('insurance.csv') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        [age, sex, bmi, children, smoker, region, charges] = row
        if line == 0:
            line += 1
            continue
        age_sum += int(age)
        regions.append(region)
        if smoker == 'yes':
            cost_smoker += float(charges)
        else:
            cost_non_smoker += float(charges)
        if int(children) > 0:
            age_sum_with_children += int(age)
        line += 1
avg_age = round(age_sum/line, 1)
regions.sort(reverse=True)
major_region = regions[0]
cost_smoker = round(cost_smoker/line, 1)
cost_non_smoker = round(cost_non_smoker/line, 1)
avg_age_with_children = round(age_sum_with_children/line, 1)
print ('average age: ', avg_age)
print ('major region: ', major_region)
print (f"cost of smokers: {cost_smoker}, cost of non smokers: {cost_non_smoker}")
print ('average age who has children: ', avg_age_with_children)
```
